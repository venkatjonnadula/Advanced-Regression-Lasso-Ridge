### Lasso Regression

```python
from sklearn.linear_model import LinearRegression, Ridge, Lasso
alphas = [0.0001, 0.001, 0.01, 0, 0.05, 0.1, 
  0.5, 1.0, 5.0, 10.0 ]


def test_alpha(a):
    model_lasso = Lasso(alpha=a)
    model_lasso.fit(X_train, y_train) 
    pred_test_lasso = model_lasso.predict(X_test)
    new_score = r2_score(y_test, pred_test_lasso)
    new_mse = mean_squared_error(y_test, pred_test_lasso)
    print('ALPHA: {:.4f} R2 SCORE: {:.4f}% new_score, {:.1f}'.format(a, new_score, new_mse))
    
    
for alpha in alphas:
    test_alpha(alpha)
    
```

At hyperparameter 0.0001 the Lasso model perform better with high R2 Score of 0.86

```python
# Gridsearch Operation on Training data set
# Objective - Find optimal value of alpha
from sklearn.model_selection import KFold
from sklearn.model_selection import GridSearchCV

folds = KFold(n_splits=10,shuffle=True,random_state=100)

hyper_param = {'alpha':[0.0001, 0.001, 0.01, 0.1,1.0, 5.0, 10.0,20.0]}

model = Lasso()

model_cv = GridSearchCV(estimator = model,
                        param_grid=hyper_param,
                        scoring='r2',
                        cv=folds,
                        verbose=1,
                        return_train_score=True
                       )

model_cv.fit(X_train,y_train)
```

```python
cv_result_train_lasso = pd.DataFrame(model_cv.cv_results_)
cv_result_train_lasso['param_alpha'] = cv_result_train_lasso['param_alpha'].astype('float32')
cv_result_train_lasso.head()
```

```python
print('For Lasso, the Best Alpha value = ', model_cv.best_params_['alpha'])
```

```python
#Lasso
import sklearn.metrics as metrics
lm = Lasso(alpha=0.0001)
lm.fit(X_train,y_train)

# train score
y_train_pred = lm.predict(X_train)
print(metrics.r2_score(y_true=y_train,y_pred=y_train_pred))

# test score
y_test_pred  = lm.predict(X_test)
print(metrics.r2_score(y_true=y_test,y_pred=y_test_pred))

new_score = r2_score(y_test, y_test_pred)
new_mse = mean_squared_error(y_test, y_test_pred)
print('ALPHA: {:.4f} R2 SCORE: {:.4f}% MSE: {:.1f}'.format(0.0001, new_score, new_mse))
```

```python
lasso_coeff = [round(x,3) for x in lm.coef_]
lasso_feature_coeff_list = list(zip(X_test.columns, lasso_coeff))

sorted(lasso_feature_coeff_list, key=lambda x: abs(x[1]), reverse=True)[:10]
```
